````python
from statsmodels.tsa.statespace.sarimax import SARIMAX
import pmdarima as pm

# Загрузка данных
data = pd.read_excel('/content/students_data.xlsx', sheet_name='data', index_col='month', parse_dates=True)

# Используем auto_arima для автоматического подбора параметров
model_auto = pm.auto_arima(data['total_num_students'], seasonal=True, m=12, trace=True)

# Обучаем модель SARIMA с подобранными параметрами
model = SARIMAX(data['total_num_students'],
                order=model_auto.order,
                seasonal_order=model_auto.seasonal_order)
model_fit = model.fit(disp=False)

# Прогноз на 12 месяцев вперед
forecast = model_fit.get_forecast(steps=12)
forecast_values = forecast.predicted_mean
forecast_ci = forecast.conf_int()  # доверительный интервал прогноза

# Вывод прогноза
print(forecast_values)
````
